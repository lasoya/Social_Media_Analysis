# Data Cleaning

## Helper Functions

```python
# identify columns to drop
remove_cols = ['cashtags', 'link', 'location', 'mentions', 'photos', 'place', 'profile_image_url', 'quote_url', 'retweet', 'urls', 'video']
```

```python
# helper function to look at value counts for each dataframe
def value_counts(dataframe):
    for column in dataframe.columns:
        print('{} \n'.format(column), dataframe[column].value_counts(), '\n \n')
```

```python
# helper function to drop columns in dataframes
def drop_columns(dataframe, cols):
    dataframe = dataframe.drop(cols, axis=1)
    return dataframe
```

# VADER

## Sentiment Intensity Analyzer

```python
# create an instance of the Sentiment Intensity Analyzer from NLTK Vader
sid = SentimentIntensityAnalyzer()
```

```python
print(sid.polarity_scores(anxiety.tweet[0]))
```

## Hashtag = Anxiety

```python
# use Vader's Sentiment Intensity Analyzer to find the sentiment for each tweet
anxiety['polarity'] = anxiety.tweet.apply(lambda x: sid.polarity_scores(x))
```

```python
# split the polarity column of dictionaries into separate columns and drop unnecessary columns
anxiety = pd.concat([anxiety.drop('polarity', axis=1), anxiety['polarity'].apply(pd.Series).drop(['neg', 'neu', 'pos'], axis=1)], axis=1)
```
