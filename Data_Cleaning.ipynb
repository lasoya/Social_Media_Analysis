# Data Cleaning

## Helper Functions

```python
# identify columns to drop
remove_cols = ['cashtags', 'created_at', 'link', 'location', 'mentions', 'photos', 'place', 'profile_image_url', 'quote_url', 'retweet', 'timezone', 'urls', 'video']
```

```python
# helper function to look at value counts for each dataframe
def value_counts(dataframe):
    for column in dataframe.columns:
        print('{} \n'.format(column), dataframe[column].value_counts(), '\n \n')
```

```python
# helper function to drop columns in dataframes
def drop_columns(dataframe, cols):
    dataframe = dataframe.drop(cols, axis=1)
    return dataframe
```

```python
# helper function to clean time column
def clean_time(time):
    time = re.sub('(:\d+)$', '', x)
    time = x.replace(':', '')
    return time
```

```python
# function to bin time into AM, PM for next cell
def time_of_day(time):
    if time < 1200:
        return 'AM'
    else:
        return 'PM'
```

```python
# function to convert month to season
def season(month):
    if month >= 3 and month < 6:
        return 'spring'
    elif month >= 6 and month < 9:
        return 'summer'
    elif month >= 9 and month < 12:
        return 'autumn'
    else:
        return 'winter'
```

```python
# check nulls
anxiety.isnull().sum()
```

```python
anxiety.columns
```
